Here we will work on entity resolution. We are moving the featured artists within the "artist" column into a new column called "FeaturedArtists" within the Charts database. 

```python
def add_featured_artists(x):
  output = 'N/A'
  if 'Featuring' in x:
    output = x[x.index('Featuring'):]
  elif '(feat.' in x:
    output = x[x.index('(feat'):]
  elif '(Feat.' in x:
    output = x[x.index('(Feat'):]
  elif '&' in x:
    output = x[x.index('&'):]
  return output
```

```python
def remove_featured_artists(x):
  if 'Featuring' in x:
    x = x[:x.index('Featuring') - 1]
  elif '(feat.' in x:
    x = x[:x.index('(feat') - 1]
  elif '(Feat.' in x:
    x = x[:x.index('(Feat') - 1]
  elif '&' in x:
    x = x[:x.index('&')-1]
  return x
```

```python
charts_df['FeaturedArtists'] = charts_df['artist'].apply(add_featured_artists)
charts_df['artist'] = charts_df['artist'].apply(remove_featured_artists)
```

```python
# Replace null values
charts_df["last-week"].fillna(0, inplace = True)
```

Here we will do a similar thing for the Spotify database. We are moving the featured artists within the "track_name" column into a new column called "FeaturedArtists" within the Spotify database. 

```python
def add_featured_artists(x):
  output = 'N/A'
  if '(feat.' in x:
    output = x[x.index('(feat.'):]
  elif '(featuring' in x:
    output = x[x.index('(featuring'):x.index(')') + 1]
  elif 'Featuring' in x:
    output = x[x.index('Featuring'):]
  return output
```

```python
def remove_featured_artists(x):
  if '(feat.' in x:
    x = x[:x.index('(feat.') - 1]
  elif '(featuring' in x:
    x = x[:x.index('(featuring') - 1]
  elif 'Featuring' in x:
    x = x[:x.index('Featuring') - 1]
  return x
```

```python
spotify_df['featured_artists'] = spotify_df['track_name'].apply(add_featured_artists)
spotify_df['track_name'] = spotify_df['track_name'].apply(remove_featured_artists)
```
